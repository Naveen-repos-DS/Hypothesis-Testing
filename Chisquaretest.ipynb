Background:
Mizzare Corporation has collected data on customer satisfaction levels for two types of smart home devices: Smart Thermostats and Smart Lights. They want to determine if there's a significant association between the type of device purchased and the customer's satisfaction level.
Data Provided:
The data is summarized in a contingency table showing the counts of customers in each satisfaction level for both types of devices:

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Satisfaction" : ['Very Satisfied','Satisfied','Neutral','Unsatisfied','Very Unsatisfied','Total'],
                   "Smart Thermostat": [50,80,60,30,20,240],
                   "Smart Light": [70,100,90,50,50,360],
                   "Total" : [120,180,150,80,70,600]})
df
```

**State the Hypotheses**

H0 - There is no significant association between the Smart Thermostats and Smart Lights.

H1 - There is a significant association between the Smart Thermostats and Smart Lights.

**Compute the Chi-Square Statisti**

```python
from scipy import stats
chi,p,dof,exp_freq = stats.chi2_contingency(df.iloc[:5,1:3])

print("Chi-Square Statistic:", chi)
print("P-Value:", p)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", exp_freq)
```

**Determine the Critical Value**

```python
alpha = 0.05
critical = stats.chi2.ppf(1-alpha,df=dof)
critical
```

```python
if chi > critical:
    print("Reject Null Hypothesis, there is a significant association between Smart Thermostats and Smart Lights.")
else:
    print("Fail to Reject Null Hypothesis, there is no significant association between Smart Thermostats and Smart Lights.")
```

**Conclusion**

Since it falls under the Null hypothesis, The  customer satisfaction levels does not depend on the Smart Thermostats and Smart Lights.
